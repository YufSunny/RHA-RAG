```python
from zai import ZhipuAiClient
import os
# 初始化客户端
client = ZhipuAiClient(api_key=os.getenv("ZHIPUAI_API_KEY"))

image_url = "https://cdn.bigmodel.cn/static/logo/introduction.png"

# 调用布局解析 API
response = client.layout_parsing.create(
    model="glm-ocr",
    file=image_url
)

# 输出结果
print(response.md_results)
```
